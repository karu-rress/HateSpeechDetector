```python
# ../HateSpeech/korean_hate_speech/news_title

import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.cluster import KMeans
from pyclustering.cluster import kmedoids
# from ckonlpy.tag import Twitter # Twitter: obsolete
from konlpy.tag import Okt
```

```python
with open("../HateSpeech/korean_hate_speech/news_title/train.news_title.txt") as f:
    lines = f.read().splitlines()

table = str.maketrans({
    '[': '',
    ']': '',
    '"': '',
    '\'': '',
})
lines = [x.translate(table) for x in lines]
lines = [x.replace('단독', '') for x in lines]
lines = [x.replace('충격', '') for x in lines]
lines = [x.replace('종합', '') for x in lines]

twitter = Okt()
# twitter.add_dictionary(['명사'], 'Noun')

titles = []
for i in lines:
    titles.append(twitter.nouns(i))

text = [' '.join(titles[i]) for i in range(len(titles))]
text
```

```python
cv = CountVectorizer()
cv.fit(text)
x = cv.transform(text)
#cv.transform(text)
```

```python
new_df = pd.DataFrame(x.toarray())

kmeans = KMeans(n_clusters=10).fit(new_df)
kmeans.labels_
```

```python
import numpy as np

lines = [x.replace('\"', '') for x in lines]

new_title = np.column_stack([kmeans.labels_, lines])
sorted_list = sorted(new_title, key=lambda x: x[0])

import csv
with open("sorted.csv", "w", newline="") as f:
    writer = csv.writer(f)
    writer.writerows(sorted_list)

with open("origin.csv", "w", newline="") as f:
    writer = csv.writer(f)
    writer.writerows(new_title)
```
